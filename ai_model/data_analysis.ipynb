```python
import pandas as pd #Библиотека для работы с данными
import seaborn as sns #Библиотека для визуализации данных
```

```python
vacancies_data = pd.read_json('vacancies.json') #Чтение данных из файла vacancies.json
```

```python
vacancies_data.head(10) #Вывод первых 10 строк данных
```

```python
vacancies_items = pd.json_normalize(vacancies_data['items']) #Нормализация данных из столбца 'items'
vacancies_items.head(10) #Вывод первых 10 строк нормализованных данных
```
